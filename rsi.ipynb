```python
df = yf.download("SPY", start="2000-01-01", end="2025-12-01")
```

```python
# display(df.de)
display(df.describe())
```

```python
plt.plot(df["Close"])
plt.title("SPY Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=50)
plt.show()
```

```python
# moving average
df["MA"] = df["Close"].rolling(window=50).mean()
```

```python
display(df.describe())
```

```python
plt.plot(df["Close"], label="Close")
plt.plot(df["MA"], label="MA")
plt.title("SPY Stock Price -- Close vs MA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=50)
plt.legend()
plt.show()
```

```python
# daily price change
df["Price_Change"] = df["Close"].diff()
```

```python
df.tail(10)
```

```python
# gain and loss
df["Gain"] = df["Price_Change"].apply(lambda x: x if x > 0 else 0)
df["Loss"] = df["Price_Change"].apply(lambda x: abs(x) if x < 0 else 0)
```

```python
df.tail(5)
```

```python
# average gain and loss
rsi_period = 14
df["Avg_Gain"] = df["Gain"].ewm(span=rsi_period, adjust=False).mean()
df["Avg_Loss"] = df["Loss"].ewm(span=rsi_period, adjust=False).mean()
```

```python
df.tail(5)
```

```python
display(df.describe())
```

```python
# calculate the Relative Strength
df["RS"] = df["Avg_Gain"] / df["Avg_Loss"]
df.head(10)
```

```python
# df["MA"].notna().sum()
# MA contain not NA values !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! only the first 50 rows contain NA
```

```python
# calculate the Relative Strength index (RSI)
# we use the RSI maths formula..... 
df["RSI"] = 100 - (100 / (1 + df["RS"]))
```

```python
display(df.describe())
```

```python
# visualization ...

plt.figure(figsize=(12, 6))
plt.plot(df["RSI"], label="RSI")
plt.title("SPY Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI value")
plt.xticks(rotation=50)
plt.axhline(70, linestyle="--", alpha=0.6, color="red", label="Overbought (70)")
plt.axhline(30, linestyle="--", alpha=0.6, color="green", label="Overbought (70)")
plt.legend()
plt.show();
```
